Чтение текста из файла

```python
def read_text(name):
    try:
        f=open(name, 'r')
        t=f.readlines()
        f.close()
        return t
    except:
        return "Ошибка при чтении файла"
```

Токенизация текста

```python
import nltk
import string

def tokenization(l):
    
    # Убираем цифры
    for d in '1234567890':
        l=l.replace(d, '')
    # Разделяем текст на слова и знаки препинания
    tokens = nltk.word_tokenize(l)
    # Избавляемся от знаков препринания и чисел. Используем генераторы
    tokens = [i for i in tokens if ( i not in string.punctuation )]
    tokens = [i for i in tokens if ( i not in {"«", "»", "\"", "``", "''", "–", "..", "..."} )]
    return tokens
```

Лемматизация текста

```python
import pymorphy2

def lemmatization(l):
    pymorph = pymorphy2.MorphAnalyzer()
    a=[]
    for word in l:
        a.append(pymorph.parse(word)[0].normal_form)
    return a
```

Подсчет частоты слов

```python
def words_frequency(words):
    unique_words = set(words)
    frequency_words = {}
    for word in unique_words:
        frequency_words[word] = words.count(word)
    return frequency_words
```

Сортировка слов в порядке убывания

```python
import operator
def sort(frequency):
    s = sorted(frequency.items(), key=operator.itemgetter(1), reverse=True)
    return s
```

Основная программа

```python
#"C:\\Users\\Mariia\\Desktop\\Homeworks_3_cource\input2.txt"
print("Введите путь к файлу:")
s=read_text(input())
if (s!="Ошибка при чтении файла"):
    text=" ".join(s).lower() 
    words=tokenization(text)
    words=lemmatization(words)
    frequency_words=words_frequency(words)
    sorted_frequency=sort(frequency_words)
    print(sorted_frequency)
else:
    print(s)
```

Тестирование функции токенизации текста

```python
class TokenozationTest(unittest.TestCase):
    def test_empty(self):
        self.assertEqual(tokenization(""), [])
    def test_only_punctuation(self):
        self.assertEqual(tokenization(" !?, ). &"), [])
    def test_without_pause(self):
        self.assertEqual(tokenization(" fff,ghh.tr!te"), ['fff', 'ghh.tr', 'te'])
    def test_other_symbols(self):
        self.assertEqual(tokenization("3 \"\", кошка, - это животное... .. * ?? ..... !!!!)) 0"), ['кошка', 'это', 'животное'])
    def test_big_text(self):
        self.assertEqual(tokenization("Существует множество приложений, позволяющих автоматизировать процессы кинотеатров, "+ 
                                      "связанные с реализацией билетов. К таким системам относятся АИС \"СУПЕРБИЛЕТ-КИНО\"[7],"), 
                                      ["Существует", 'множество', "приложений", 'позволяющих', 'автоматизировать', 'процессы',
                                       "кинотеатров", 'связанные', 'с', 'реализацией', 'билетов', 'К', 'таким', 'системам',
                                       'относятся', 'АИС', 'СУПЕРБИЛЕТ-КИНО'])
    def test_another_value(self):        
        with self.assertRaises(AttributeError):
            tokenization(3)
suite = unittest.TestLoader().loadTestsFromTestCase(TokenozationTest)
unittest.TextTestRunner(verbosity=2, stream=sys.stdout).run(suite) 
```

Тестирование функции лемматизации текста

```python
class LemmatizationTest(unittest.TestCase):
    def test_empty(self):
        self.assertEqual(lemmatization(""), [])
    def test_another_value(self):        
        with self.assertRaises(TypeError):
            lemmatization(3)
    def test_normal_form(self):
        self.assertEqual(lemmatization(["она", "лежать", "на", "холодный", "земля"]), ["она", "лежать", "на", "холодный", "земля"])
    def test_npt_normal_form(self):
        self.assertEqual(lemmatization(["Она", "лежит", "на", "холодной", "земле"]), ["она", "лежать", "на", "холодный", "земля"])
    def test_npt_normal_form(self):
        self.assertEqual(lemmatization(["Она", "лежит", "на", "холодной", "земле", ",", "yours"]), ["она", "лежать", "на", "холодный", "земля", ",", "yours"])
    def test_big_text(self):
        self.assertEqual(lemmatization(["Существует", 'множество', "приложений", 'позволяющих', 'автоматизировать', 'процессы',
                                       "кинотеатров", 'связанные', 'с', 'реализацией', 'билетов', 'К', 'таким', 'системам',
                                       'относятся', 'АИС', 'СУПЕРБИЛЕТ-КИНО']), 
                                       ["существовать", 'множество', "приложение", 'позволять', 'автоматизировать', 'процесс',
                                       "кинотеатр", 'связанный', 'с', 'реализация', 'билет', 'к', 'такой', 'система',
                                       'относиться', 'аиса', 'супербилет-кино'])
    def test_several_variants(self):
        self.assertEqual(lemmatization(["делала","из",'стали']), ["делать","из",'стать'])
suite = unittest.TestLoader().loadTestsFromTestCase(LemmatizationTest)
unittest.TextTestRunner(verbosity=2, stream=sys.stdout).run(suite) 
```

Тестирование функции подсчета частоты слов

```python
class WordsFrequencyTest(unittest.TestCase):
    def test_empty(self):
        self.assertEqual(words_frequency([]), {})
    def test_another_value(self):        
        with self.assertRaises(TypeError):
            words_frequency(3)
    def test_frequency_1(self):
        self.assertDictEqual(words_frequency(['a', 'b', 'c']), {'a': 1, 'b': 1, 'c': 1})
    def test_frequency_more_than_1(self):
        self.assertDictEqual(words_frequency(['a', 'b', 'c', 'b', 'b', 'c']), {'b': 3, 'c': 2, 'a': 1})

suite = unittest.TestLoader().loadTestsFromTestCase(WordsFrequencyTest)
unittest.TextTestRunner(verbosity=2, stream=sys.stdout).run(suite)
```

Тестирование функции сортировки

```python
class SortTest(unittest.TestCase):
    def test_empty(self):
        self.assertEqual(sort({}), [])
    def test_another_value(self):        
        with self.assertRaises(AttributeError):
            sort(3)
    def test_another_dict(self):        
        with self.assertRaises(AttributeError):
            sort({"5"})
    def test_one_element(self):
        self.assertEqual(sort({"5":7}), [("5",7)])
    def test_sorted(self):
        self.assertEqual(sort({"5":7, "4":5, "9":2, "1":1 }), [("5",7), ("4",5), ("9",2), ("1",1)])
    def test_sorted_reverce(self):
        self.assertEqual(sort({"5":1, "4":2, "9":5, "1":7 }), [("1",7), ("9",5), ("4",2), ("5",1)])
    def test_not_sorted(self):
        self.assertEqual(sort({"5":8, "4":1, "9":1, "1":2 }), [('5', 8), ('1', 2), ('4', 1), ('9', 1)])

suite = unittest.TestLoader().loadTestsFromTestCase(SortTest)
unittest.TextTestRunner(verbosity=2, stream=sys.stdout).run(suite)
```
