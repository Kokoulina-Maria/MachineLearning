Чтение текста из файла

```python
def read_text(name):
    try:
        f=open(name, 'r')
        t=f.readlines()
        f.close()
        return t
    except:
        return "Ошибка при чтении файла"
```

Токенизация текста

```python
import nltk
import string

def tokenization(l):
    
    # Убираем цифры
    for d in '1234567890\'/':
        l=l.replace(d, ' ')
    # Разделяем текст на слова и знаки препинания
    tokens = nltk.word_tokenize(l)
    # Избавляемся от знаков препринания и чисел. Используем генераторы
    tokens = [i for i in tokens if ( i not in string.punctuation )]
    tokens = [i for i in tokens if ( i not in {"«", "»", "\"", "``", "''", "–", "..", "...", "‘", "’"} )]
    return tokens
```

Лемматизация текста

```python
import pymorphy2

def lemmatization(l):
    pymorph = pymorphy2.MorphAnalyzer()
    a=[]
    for word in l:
        a.append(pymorph.parse(word)[0].normal_form)
    return a
```

Подсчет частоты слов

```python
def words_frequency1(words, l):
    unique_words = set(words)
    frequency_words = {}
    for word in unique_words:
        frequency_words[word] = round(words.count(word)/l, 5)
    return frequency_words

def words_frequency(words):
    unique_words = set(words)
    frequency_words = {}
    for word in unique_words:
        frequency_words[word] = words.count(word)
    return frequency_words
```

Сортировка слов в порядке убывания

```python
import operator
def sort(frequency):
    s = sorted(frequency.items(), key=operator.itemgetter(1), reverse=True)
    return s
```

## TF-IDF по предложениям

TF было найдено ранее
IDF ищем по предложениям.

Разбиваем текст на предложения:

```python
import re
sentences=re.split(r'(?<=[.!?…]) ', text)
```

Сколько предложений всего:

```python
count_sen=len(sentences)
```

Каждое предложение в тексте мы токенизируем и лемматизируем:

```python
#список для леммат. предложений
lem_sent=[]

for sentence in sentences:
    w=lemmatization(tokenization(sentence))
    lem_sent.append(w)
```

Теперь перебираем все слова из frequency_words1. Записываем в отдельный словарь их IDF.

```python
import math
idf={}
for word in sorted_frequency1:
    i=0
    for sentence in lem_sent:
        for w in sentence:
            if w==word[0]:
                i=i+1
                break
    if (i!=0):
        idf[word[0]] = round(math.log10(count_sen/i), 3)
    else:
        idf[word[0]]=0
```

Далее умножаем TF на IDF и записываем в отдельный словарь:

```python
tf_idf={}
for word in sorted_frequency1:
    tf_idf[word[0]]=round(word[1]*idf[word[0]],5)
```

Сортируем по убыванию:

```python
tf_idf_sorted = sorted(tf_idf.items(), key=operator.itemgetter(1), reverse=True)
```

Удаляем слова с наибольшим показателем. Как видим, не все стоп-слова оказались в этом списке в начале. Кроме того, многие важные слова, типо "мяч, текст, термин, количество" приходится удалить =( Кроме того, для всех текстов граница, до которой все элементы слудует удалить, разная

```python
i=0
while i< len(tf_idf_sorted):
    if (tf_idf_sorted[i][1]>0.00545):
        del tf_idf_sorted[i]
    else:
        i=i+1
print(tf_idf_sorted)
```
